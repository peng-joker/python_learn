# 执行ChatOpenAi

大语言模型：定义一个ChatOpenAI对象，model可以选择支持的大语言模型名称

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="glm-4")
```

先尝试直接调用一下

```python
from langchain_core.messages import HumanMessage, SystemMessage
messages = "请把以下内容翻译成中文：hi"
result = model.invoke(messages)
result
```

OutputParsers：StrOutputParser获取输出

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

定义链，并执行

使用 | 运算符来轻松地创建链。| 运算符用于在LangChain中将两个元素组合在一起。

```python
chain = model | parser
chain.invoke(messages)
```

提示词模板：定义模板

```python
from langchain_core.prompts import ChatPromptTemplate
system_template = "将以下内容翻译成{language}，只需要输出翻译后的内容:"
prompt_template = ChatPromptTemplate.from_messages(
[("system", system_template), ("user", "{text}")]
)
```

打印模板信息

```python
result = prompt_template.invoke({"language": "中文", "text": "hi"})
result.to_messages()
```

定义链

```python
chain = prompt_template | model | parser
```

执行

```python
chain.invoke({"language": "中文", "text": "Applications that can reason. Powered by LangChain."})
```
