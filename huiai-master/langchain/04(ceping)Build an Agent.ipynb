```python
import os
from google.colab import userdata
os.environ["TAVILY_API_KEY"] = userdata.get('TAVILY_API_KEY')
from langchain_community.tools.tavily_search import TavilySearchResults
search = TavilySearchResults(max_results=2)
search.invoke("what is the weather in beijing")
```

## æœ¬åœ°çŸ¥è¯†åº“æ£€ç´¢å·¥å…·

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
# åŠ è½½ç½‘é¡µå†…å®¹
loader = WebBaseLoader("https://docs.smith.langchain.com/overview")
docs = loader.load()
# åˆ†å‰²æ–‡æœ¬
documents = RecursiveCharacterTextSplitter(
    chunk_size=1000, chunk_overlap=200
).split_documents(docs)
# ç”ŸæˆåµŒå…¥å‘é‡å¹¶å­˜å‚¨åœ¨FAISSå‘é‡æ•°æ®åº“ä¸­
vector = FAISS.from_documents(documents, OpenAIEmbeddings(openai_api_base=userdata.get('OPENAI_API_BASE'),openai_api_key=userdata.get('OPENAI_API_KEY')))
retriever = vector.as_retriever()
retriever.invoke("how to upload a dataset")[0]
```

å°†retrieverå°è£…ä¸ºä¸€ä¸ªå·¥å…·ä¾›æ™ºèƒ½ä½“è°ƒç”¨

```python
from langchain.tools.retriever import create_retriever_tool
retriever_tool = create_retriever_tool(
    retriever,
"langsmith_search",
"Search for information about LangSmith. For any questions about LangSmith, you must use this tool!",
)
```

## åˆ›å»ºå·¥å…·é›†

```python
tools = [search, retriever_tool]
for t in tools:
  print("--")
  print(t.name)
  print(t.description)
  print(t.args)
```
